```python
import math
from scipy.stats import norm
```

```python
# population data
population_mean = 500
population_std = 40
```

```python
# sample data from farms
sample_mean = 470
sample_size = 36
```

```python
# significance level
alpha = 0.05
```

```python
# calculate z-score
z_score = (sample_mean - population_mean) / (
    population_std / math.sqrt(sample_size)
)
```

```python
# calculate p-value for left-tailed test
p_value = norm.cdf(z_score)
```

```python
# display project results
print("------ smart farming z-test ------")
print("old average water usage:", population_mean)
print("new average water usage:", sample_mean)
print("z-score:", round(z_score, 2))
print("p-value:", round(p_value, 5))
```

```python
# hypothesis decision
if p_value < alpha:
    print("result: reject the null hypothesis")
    print("the new irrigation system significantly reduces water usage")
else:
    print("result: fail to reject the null hypothesis")
    print("there is no significant reduction in water usage")
```
