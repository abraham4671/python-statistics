```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
```

```python
np.random.seed(42)

old_scores = np.random.normal(loc=30, scale=3, size=30)
new_scores = np.random.normal(loc=40, scale=4, size=30)

data = pd.DataFrame({
    "group": ["old"] * 30 + ["new"] * 30,
    "workout_minutes": list(old_scores) + list(new_scores)
})

print(data.head())
```

```python
# separate groups
old_group = data[data["group"] == "old"]["workout_minutes"]
new_group = data[data["group"] == "new"]["workout_minutes"]
```

```python
# perform independent t-test
t_statistic, p_value = ttest_ind(old_group, new_group)
```

```python
# significance level
alpha = 0.05
```

```python
# display results
print("T-Statistic:", round(t_statistic, 3))
print("P-Value:", round(p_value, 5))
```

```python
# hypothesis decision
if p_value < alpha:
    print("Reject Null Hypothesis")
    print("The new recommendation system improved workout time.")
else:
    print("Fail to Reject Null Hypothesis")
    print("No significant improvement detected.")
```
