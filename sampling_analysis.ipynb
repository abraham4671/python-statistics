```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
population = np.random.normal(loc=50, scale=10, size=10000)

print(population[:10])
```

```python
sns.histplot(population, kde=True)
plt.title("Population Distribution")
plt.show()
```

```python
sample_random = np.random.choice(population, size=100, replace=False)

print(sample_random)
```

```python
step = len(population) // 100

systematic_sample = population[::step]

print(systematic_sample[:10])
```

```python
step = len(population) // 100

systematic_sample = population[::step]

print(systematic_sample[:10])
```

```python
data = pd.DataFrame({
    'Marks': population
})

data['Category'] = pd.cut(
    data['Marks'],
    bins=[0, 40, 60, 100],
    labels=['Low', 'Medium', 'High']
)

stratified_sample = data.groupby('Category').sample(30)

print(stratified_sample.head())
```

```python
clusters = np.array_split(population, 10)

cluster_sample = clusters[3]

print(cluster_sample[:10])
```

```python
x = np.linspace(20, 80, 100)

pdf = norm.pdf(x, loc=50, scale=10)

plt.plot(x, pdf)
plt.title("Normal Distribution Curve")
plt.show()
```

```python
sample_means = []

for i in range(1000):
    sample = np.random.choice(population, size=30)
    sample_means.append(np.mean(sample))

sns.histplot(sample_means, kde=True)
plt.title("Distribution of Sample Means")
plt.show()
```

```python
methods = ['Random', 'Systematic', 'Stratified']

means = [
    np.mean(sample_random),
    np.mean(systematic_sample),
    np.mean(stratified_sample['Marks'])
]

plt.bar(methods, means)
plt.title("Sampling Technique Comparison")
plt.ylabel("Mean")
plt.show()
```
